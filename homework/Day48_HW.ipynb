```python
import numpy as np 
import pandas as pd
#from sklearn.metrics import accuracy_score 
#from sklearn.model_selection import cross_val_score
import os

print(os.listdir("./inputs_day48"))
```

```python
train_data = pd.read_csv('./inputs_day48/train.csv',header = None)
train_labels = pd.read_csv('./inputs_day48/trainLabels.csv',header = None)
test_data =  pd.read_csv('./inputs_day48/test.csv',header = None)

train_data.head()
```

```python
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
x_train,x_test,y_train,y_test = train_test_split(train_data,train_labels, test_size = 0.30, random_state = 42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import GradientBoostingClassifier
import numpy as np

y_train_d1 = y_train[0][:,np.newaxis]
clf = GradientBoostingClassifier()
clf.fit(x_train, y_train_d1.ravel())
y_pred = clf.predict(x_test)
y_pred
```

```python
from sklearn import metrics

acc = metrics.accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)
print("Feature importance: ", clf.feature_importances_)
```

```python
n_estimators = [100, 200, 300, 400]
max_depth = [1, 3, 5, 7, 9]
param_grid = dict(n_estimators=n_estimators, max_depth=max_depth)
grid_search = GridSearchCV(clf, param_grid, scoring="accuracy", n_jobs=-1, verbose=1)
grid_result = grid_search.fit(x_train, y_train_d1.ravel())
```

```python
print("Best Accuracy: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
clf_bestparam = GradientBoostingClassifier(max_depth=grid_result.best_params_['max_depth'],
                                           n_estimators=grid_result.best_params_['n_estimators'])
clf_bestparam.fit(x_train, y_train_d1.ravel())
y_pred2 = clf_bestparam.predict(x_test)
acc2 = metrics.accuracy_score(y_test, y_pred2)
print("Accuracy: ", acc2)
print("Feature importance: ", clf_bestparam.feature_importances_)
```

```python
y_pred3 = clf_bestparam.predict(test_data)
df_pred = pd.DataFrame(y_pred3)
df_pred.index += 1
df_pred.columns = ['Solution']
df_pred['Id'] = np.arange(1,df_pred.shape[0]+1)
df_pred = df_pred[['Id', 'Solution']]
df_pred.head()
```

```python
df_pred.to_csv('Submission_GBDT.csv',index=False)
```
